```python
from wordcloud import WordCloud
```

```python
f = open(r'C:\Users\cqiuac\Documents\WeChat Files\choryat\FileStorage\File\2019-05\2018.txt','r').read()
wordcloud = WordCloud(background_color="white",width=1000, height=860, margin=2).generate(f)
import matplotlib.pyplot as plt
plt.imshow(wordcloud)
plt.axis("off")
plt.show()

wordcloud.to_file(r'C:\Users\cqiuac\PycharmProjects\trying\AI_in\sensetime_2018.png')
```
